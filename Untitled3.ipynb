```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
trades = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed_index.csv")
```

```python
print(trades.columns)
print(trades.head(), '\n')
```

```python
print(sentiment.columns)
print(sentiment.head(), '\n')
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
sentiment = sentiment.rename(columns={'classification': 'Sentiment'})
```

```python
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], errors='coerce', format='%d-%m-%Y %H:%M')
trades['date'] = trades['Timestamp IST'].dt.date
```

```python
merged = pd.merge(trades, sentiment[['date', 'Sentiment']], on='date', how='inner')
```

```python
merged['Closed PnL'] = pd.to_numeric(merged['Closed PnL'], errors='coerce')
merged['Fee'] = pd.to_numeric(merged['Fee'], errors='coerce')
```

```python
print("\nTrades per Sentiment:")
print(merged['Sentiment'].value_counts())

print("\nAverage Closed PnL by Sentiment:")
print(merged.groupby('Sentiment')['Closed PnL'].mean().sort_values(ascending=False))
```

```python
sns.boxplot(x='Sentiment', y='Closed PnL', data=merged)
plt.title("Closed PnL Distribution by Market Sentiment")
plt.xticks(rotation=45)
plt.axhline(0, color='gray', linestyle='--')
plt.show()
```

```python
sns.countplot(data=merged, x='Sentiment', hue='Direction')
plt.title("Buy/Sell Direction under Market Sentiment")
plt.xticks(rotation=45)
plt.show()
```

```python
merged['Size USD'] = pd.to_numeric(merged['Size USD'], errors='coerce')
sns.barplot(data=merged, x='Sentiment', y='Size USD', estimator='mean', ci=None)
plt.title("Average Trade Size (USD) by Sentiment")
plt.xticks(rotation=45)
plt.show()
```
